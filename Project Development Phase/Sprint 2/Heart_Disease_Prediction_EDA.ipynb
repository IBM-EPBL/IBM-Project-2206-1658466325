4) Data Description

The dataset has 270 records and 14features. The last column predicts whether a person has heart disease or not. The detailed description of dataset is given below,

1). Age - The age of the patient varies in the range of 40-80.

2). Sex - It describes the gender of the user.

0 - Female;
1 - Male

3). Chest Pain - The chest pain is categorized into four types as follows,

 1 - Typical angina;
 2 - Atypical angina;
 3 - Nonanginal pain;
 4 -Asymptomatic

4). BP - It describes the blood pressure level in the range of 100-800.

5). Cholestrol - It describes the Serum Cholestrol in milligrams/deciliters in the range of 100-600.

6). FBS over 120 - It shows the Fasting Blood Sugar(FBS) as follows,

1 - Presence of blood;
0 - Absence of blood sugar

7). EKG Results - The ElectroCardioGram(ECG) results are shown as follows,

0 - Normal;
1 - Having ST wave abnormality;
2 - Having high Abnormality;
8). Max HR - It shows maximum heart rate ith range of 100-200.

9). Exercise Angina - It shows the pain experienced by a person after doing exercises.

0 - Absence;
1 - Presence

10). ST Depression - It shows the result of ECG in the range of 0-4.

11). Slope of ST - The slope of the [eak exercise ST segment is shown in the range of 1-3.

1 - Downsloping;
2 - Flat;
3 - Upsloping

12). Number of vessels fluoro - It shows thw number of major vessels colored by fluoroscopy in the range of 0-3.

13). Thallium - It shows the blood flow in heart as follows,

0 - Normal;
1 - Fixed defect;
2 - Reversible defect

14). Heart disease - It is the target variable which predicts if a person has heart disease or not.

0 - Absence;
1 - Presence
