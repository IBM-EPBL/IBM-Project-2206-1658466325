```python
import numpy as np
for x in ['Rings']:
    q75,q25 = np.percentile(data.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    data.loc[data[x] < min,x] = np.nan
    data.loc[data[x] > max,x] = np.nan
```

```python
data = data.dropna(axis = 0)
```

```python
import numpy as np
for x in ['Shell_weight']:
    q75,q25 = np.percentile(data.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    data.loc[data[x] < min,x] = np.nan
    data.loc[data[x] > max,x] = np.nan
```

Build, train and test the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python
x = data.drop(['Sex', 'Rings'], axis = 1)
y = data.Rings
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.dropna(inplace=True)
```

```python
x_train.isnull().sum()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_train.drop(y_train.tail(20).index,inplace = True)
 
```

```python
regressor = LinearRegression()  
regressor.fit(x_train, y_train) #training the algorithm
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
x_test.dropna(inplace=True)
```

```python
x_test.isnull().sum()
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_test.shape
```

```python
y_test.drop(y_test.tail(5).index,inplace = True)
```

```python
y_pred.shape
```

```python
c = pd.DataFrame({"Actual": y_test, "Predictions": regressor.predict(x_test)})
c
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
