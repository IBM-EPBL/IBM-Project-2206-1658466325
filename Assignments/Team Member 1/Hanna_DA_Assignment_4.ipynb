```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
from sklearn import metrics
import numpy as np
```

```python
convert = StandardScaler()

feat = df.drop(['Sex', 'Rings'], axis = 1)
label = df.Rings

feat = convert.fit_transform(feat)
```

```python
f_train, f_test, l_train, l_test = train_test_split(feat, label, random_state = 23, test_size = 0.2)
```

```python
model1 = LogisticRegression(random_state=23)
model1.fit(f_train, l_train)
```

```python
y_predict = model1.predict(f_train)
```

```python
accuracy_score(l_train, y_predict) 
```
